# Natural Language Processing (NLP)

Natural Language Processing helps computers understand and work with human language.

NLP is used in spam detection, chatbots, sentiment analysis, translation, search engines, and text summarization.

# Common NLP Steps

- Text cleaning
- Lowercasing
- Tokenization
- Stopword removal
- Stemming or lemmatization
- Vectorization
- Model training

# Text Vectorization

Machine learning models need numbers, not raw text.

Common methods:

- Bag of Words
- TF-IDF
- Word embeddings
- Transformer embeddings

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
texts = [
    'I love this movie',
    'This film was amazing',
    'What a fantastic story',
    'I hated this movie',
    'This film was boring',
    'The story was terrible',
    'Amazing acting and great music',
    'Bad acting and boring scenes'
]

labels = [1, 1, 1, 0, 0, 0, 1, 0]

df = pd.DataFrame({'text': texts, 'sentiment': labels})
df
```

```python
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(df['text'])

bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())
bow_df
```

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['text'])

model = MultinomialNB()
model.fit(X_tfidf, df['sentiment'])

new_reviews = ['great movie and amazing story', 'boring film with bad acting']
new_vectors = tfidf.transform(new_reviews)
predictions = model.predict(new_vectors)

for review, pred in zip(new_reviews, predictions):
    print(review, '->', 'positive' if pred == 1 else 'negative')
```
