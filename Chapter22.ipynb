# Time Series Forecasting

Time Series Forecasting predicts future values based on time-ordered data.

Examples include sales forecasting, stock prices, weather, website traffic, and demand forecasting.

# Time Series Components

- **Trend:** long-term upward or downward movement.
- **Seasonality:** repeating patterns.
- **Noise:** random variation.
- **Cyclic patterns:** longer-term repeated movements.

# Important Rule

Never randomly shuffle time series data for forecasting. Training should use past data and testing should use future data.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
dates = pd.date_range(start='2025-01-01', periods=120, freq='D')
trend = np.linspace(50, 100, 120)
seasonality = 10 * np.sin(np.arange(120) * 2 * np.pi / 30)
noise = np.random.default_rng(42).normal(0, 3, 120)
sales = trend + seasonality + noise

df = pd.DataFrame({'date': dates, 'sales': sales})
df.head()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df['date'], df['sales'])
plt.title('Daily Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
df['day_number'] = np.arange(len(df))
df['lag_1'] = df['sales'].shift(1)
df['rolling_7'] = df['sales'].rolling(7).mean()
df = df.dropna()

train = df.iloc[:-20]
test = df.iloc[-20:]

features = ['day_number', 'lag_1', 'rolling_7']
model = LinearRegression()
model.fit(train[features], train['sales'])

preds = model.predict(test[features])

print('MAE:', mean_absolute_error(test['sales'], preds))
print('RMSE:', np.sqrt(mean_squared_error(test['sales'], preds)))
```

```python
plt.figure(figsize=(10, 4))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Actual')
plt.plot(test['date'], preds, label='Forecast')
plt.legend()
plt.title('Forecast vs Actual')
plt.show()
```
