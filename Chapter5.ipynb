# Statistics for Machine Learning

## What is Statistics?

Statistics is the science of collecting, analyzing, interpreting, and presenting data.

Machine Learning heavily relies on statistics to:

- Understand datasets
- Find patterns
- Make predictions
- Measure uncertainty
- Evaluate models

## Types of Statistics

### Descriptive Statistics

Summarizes data.

Examples:

- Mean
- Median
- Mode
- Variance
- Standard Deviation

### Inferential Statistics

Makes predictions about populations using samples.

Examples:

- Hypothesis Testing
- Confidence Intervals
- Probability Theory

# Sample vs Population

## Population

Entire dataset of interest.

Example:

All students in a university.

## Sample

Subset of the population.

Example:

100 students selected from the university.

```text
Population
 ├── Student 1
 ├── Student 2
 ├── ...
 └── Student N

Sample
 ├── Student A
 ├── Student B
 └── Student C
```

````python
import numpy as np
import pandas as pd

data = [10, 20, 30, 40, 50]

print("Dataset:", data)
````

# Mean (Average)

The mean is the average value of a dataset.

Formula:

Mean = Sum of all values / Number of values

Example:

Data = [10, 20, 30, 40, 50]

Mean = (10 + 20 + 30 + 40 + 50) / 5

Mean = 30

````python
import numpy as np

data = [10, 20, 30, 40, 50]

print("Mean:", np.mean(data))
````

# Median

Median is the middle value after sorting.

Example:

Data = [10,20,30,40,50]

Median = 30

If even numbers exist:

Data = [10,20,30,40]

Median = (20+30)/2 = 25

## Advantage

Less affected by outliers.

````python
data = [10,20,30,40,50]

print("Median:", np.median(data))
````

# Mode

Mode is the most frequently occurring value.

Example:

[10,20,20,30,40]

Mode = 20

````python
from scipy import stats

data = [10,20,20,30,40]

mode = stats.mode(data, keepdims=True)

print("Mode:", mode.mode[0])
````

# Mean vs Median vs Mode

| Measure | Description |
|----------|-------------|
| Mean | Average |
| Median | Middle Value |
| Mode | Most Frequent Value |

### Example with Outlier

Data:

[10,20,30,40,500]

Mean = 120

Median = 30

Notice how the outlier affects the mean.

# Variance

Variance measures how far values are spread from the mean.

Low Variance:

```text
10 11 12 13 14
```

High Variance:

```text
1 10 50 100 200
```

Formula:

Variance = Average of squared deviations from mean

````python
data = [10,20,30,40,50]

print("Variance:", np.var(data))
````

# Standard Deviation

Standard Deviation is the square root of variance.

Measures how spread out data is.

Low SD → Data close to mean

High SD → Data widely spread

Formula:

SD = √Variance

````python
print("Standard Deviation:", np.std(data))
````

# Probability

Probability measures the likelihood of an event.

Formula:

Probability = Favorable Outcomes / Total Outcomes

Example:

Rolling a die

Probability of getting 4:

1/6 = 0.1667

````python
probability = 1 / 6

print(probability)
````

# Conditional Probability

Probability of an event occurring given another event occurred.

Formula:

P(A|B)

Read as:

Probability of A given B

# Bayes Theorem

Bayes theorem updates probability using new evidence.

Formula:

P(A|B) = [P(B|A) × P(A)] / P(B)

Applications:

- Spam Detection
- Medical Diagnosis
- Recommendation Systems

# Normal Distribution

Most important probability distribution.

Characteristics:

- Bell-shaped curve
- Mean = Median = Mode
- Symmetrical

Examples:

- Human Height
- IQ Scores
- Exam Marks

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50,10,1000)

plt.hist(data,bins=30)

plt.title("Normal Distribution")

plt.show()
````

# Empirical Rule (68-95-99.7 Rule)

Normal Distribution:

68% of values lie within:

Mean ± 1 SD

95% of values lie within:

Mean ± 2 SD

99.7% of values lie within:

Mean ± 3 SD

# Z-Score

Measures how many standard deviations a value is from the mean.

Formula:

Z = (X - Mean) / SD

Interpretation:

Z = 0 → At mean

Z > 0 → Above mean

Z < 0 → Below mean

````python
from scipy.stats import zscore

data = [10,20,30,40,50]

print(zscore(data))
````

# Covariance

Measures how two variables move together.

Positive Covariance:

Both increase together.

Negative Covariance:

One increases while the other decreases.

````python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

print(np.cov(x,y))
````

# Correlation

Normalized version of covariance.

Range:

-1 to +1

### Values

+1 → Perfect Positive

0 → No Relationship

-1 → Perfect Negative

````python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

print(np.corrcoef(x,y))
````

# Central Limit Theorem (CLT)

One of the most important statistical concepts.

Statement:

As sample size increases, the sampling distribution of the mean approaches a normal distribution.

Even if the original population is not normally distributed.

Applications:

- Statistical Inference
- Confidence Intervals
- Hypothesis Testing

# Hypothesis Testing

Used to determine whether a claim is true.

### Null Hypothesis (H₀)

No effect exists.

### Alternative Hypothesis (H₁)

Effect exists.

Example:

H₀: New teaching method has no impact.

H₁: New teaching method improves scores.

# P-Value

P-value measures evidence against the null hypothesis.

Common threshold:

0.05

If:

P < 0.05

Reject H₀

Otherwise:

Fail to Reject H₀

# Confidence Interval

A range likely to contain the true population parameter.

Example:

95% Confidence Interval

[48, 52]

Interpretation:

We are 95% confident that the true mean lies between 48 and 52.
