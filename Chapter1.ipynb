## What is Machine Learning?

Machine Learning (ML) is a branch of Artificial Intelligence (AI) that enables computers to learn from data and improve their performance without being explicitly programmed.

### Traditional Programming

Input + Program → Output

### Machine Learning

Input + Output Data → Learning Algorithm → Model

The model learns patterns from historical data and can make predictions on unseen data.

### Applications

- Netflix Recommendations
- YouTube Suggestions
- Face Recognition
- Self-Driving Cars
- Fraud Detection
- Medical Diagnosis
- Chatbots

## Why Do We Need Machine Learning?

Traditional programming becomes difficult when:

- Data is very large
- Rules are complex
- Patterns change frequently

### Benefits

1. Automation
2. Better Predictions
3. Pattern Discovery
4. Continuous Improvement

Example:
Spam Detection systems learn automatically from emails rather than using manually written rules.

## How Machine Learning Works

### Step 1: Collect Data

Example:

| Area (sq.ft) | Price (Lakhs) |
|-------------|--------------|
| 1000 | 20 |
| 1200 | 25 |
| 1500 | 30 |

### Step 2: Data Preprocessing

- Handle Missing Values
- Remove Duplicates
- Correct Errors

### Step 3: Select Algorithm

Examples:

- Linear Regression
- Decision Tree
- Random Forest
- KNN

### Step 4: Train Model

The algorithm learns patterns from the training data.

### Step 5: Evaluate Model

Common Metrics:

- Accuracy
- Precision
- Recall
- F1 Score

### Step 6: Make Predictions

### Supervised Learning

Uses labeled data.

Examples:

- House Price Prediction
- Spam Detection

Algorithms:

- Linear Regression
- Logistic Regression
- Decision Trees
- SVM

### Unsupervised Learning

Uses unlabeled data.

Examples:

- Customer Segmentation
- Market Basket Analysis

Algorithms:

- K-Means
- PCA
- DBSCAN

### Reinforcement Learning

Learns through rewards and penalties.

Examples:

- Robotics
- Chess AI
- Self-Driving Cars

## Artificial Intelligence vs Machine Learning

````python
# Simple Machine Learning Example

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1000], [1200], [1500]])
y = np.array([20, 25, 30])

model = LinearRegression()

model.fit(X, y)

prediction = model.predict([[1800]])

print("Predicted Price:", prediction[0])
````
