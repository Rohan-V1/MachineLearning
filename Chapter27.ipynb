# Feature Stores

A Feature Store is a centralized system for storing, managing, and serving machine learning features.

It helps teams reuse features consistently across training and production.

# Why Feature Stores Are Useful

Feature stores help solve:

- Duplicate feature engineering
- Training-serving skew
- Inconsistent transformations
- Poor feature discoverability
- Slow production feature lookup

# Offline vs Online Store

- **Offline feature store:** stores historical features for training.
- **Online feature store:** serves fresh features quickly for real-time prediction.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import pandas as pd
```

```python
raw_data = pd.DataFrame({
    'customer_id': [1, 1, 2, 2, 3, 3],
    'purchase_amount': [100, 150, 50, 70, 200, 250],
    'days_since_purchase': [10, 2, 20, 5, 4, 1]
})

raw_data
```

```python
feature_table = raw_data.groupby('customer_id').agg(
    avg_purchase_amount=('purchase_amount', 'mean'),
    total_purchase_amount=('purchase_amount', 'sum'),
    min_days_since_purchase=('days_since_purchase', 'min')
).reset_index()

feature_table
```

```python
online_feature_store = feature_table.set_index('customer_id').to_dict(orient='index')

def get_customer_features(customer_id):
    return online_feature_store.get(customer_id, None)

get_customer_features(1)
```

# Advantages / Uses

# Limitations / Things to Remember
