# Model Deployment (Flask/FastAPI)

Model Deployment makes a trained model available to users or applications.

Flask and FastAPI are popular Python frameworks for serving ML models through APIs.

# Deployment Steps

1. Train the model.
2. Save the model.
3. Create an API endpoint.
4. Load the model inside the app.
5. Accept input data.
6. Return prediction.
7. Test and monitor the service.

# Flask vs FastAPI

- **Flask:** simple, flexible, widely used.
- **FastAPI:** fast, modern, automatic docs, good type validation.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import joblib
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])

pipeline.fit(X, y)
joblib.dump(pipeline, 'chapter28_iris_model.joblib')

print('Model saved')
```
