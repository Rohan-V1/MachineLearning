# ML System Design

ML System Design focuses on building complete machine learning systems, not just models.

It includes data flow, training, serving, monitoring, retraining, scalability, reliability, and business constraints.

# Core Components

An ML system usually contains:

- Data sources
- Data pipeline
- Feature engineering
- Model training
- Model registry
- Prediction service
- Monitoring
- Feedback loop

# Design Questions

Before designing an ML system, ask:

- Is prediction real-time or batch?
- What is the latency requirement?
- How often does data change?
- What metric matters?
- How will failures be handled?
- How will model drift be detected?

# Python Implementation

The following cells show a small practical example for this chapter.

```python
from dataclasses import dataclass
from typing import List
```

```python
@dataclass
class MLSystemDesign:
    problem_type: str
    prediction_mode: str
    data_sources: List[str]
    model_type: str
    evaluation_metric: str
    monitoring_signals: List[str]


fraud_system = MLSystemDesign(
    problem_type='binary classification',
    prediction_mode='real-time',
    data_sources=['transactions', 'user_profile', 'device_info'],
    model_type='gradient boosted trees',
    evaluation_metric='precision-recall AUC',
    monitoring_signals=['latency', 'data drift', 'fraud rate', 'prediction distribution']
)

fraud_system
```

```python
def estimate_daily_predictions(requests_per_second):
    return requests_per_second * 60 * 60 * 24

print('Daily predictions:', estimate_daily_predictions(50))
```

```python
def choose_serving_pattern(latency_needed, batch_allowed):
    if batch_allowed:
        return 'Batch prediction pipeline'
    if latency_needed == 'low':
        return 'Online prediction API'
    return 'Async prediction service'

choose_serving_pattern(latency_needed='low', batch_allowed=False)
```
