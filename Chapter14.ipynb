# Clustering (K-Means, Hierarchical, DBSCAN)

Clustering is an unsupervised learning technique used to group similar data points together.

Unlike classification, clustering does not use labeled target values.

# K-Means Clustering

K-Means forms K clusters by repeatedly:

1. Choosing K centroids.
2. Assigning each point to the nearest centroid.
3. Updating centroids using the mean of assigned points.
4. Repeating until clusters stabilize.

# Hierarchical and DBSCAN

**Hierarchical clustering** builds a tree-like structure of clusters.

**DBSCAN** groups points based on density and can detect noise or outliers.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1])
plt.title('Unlabeled Data')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
plt.title('K-Means Clustering')
plt.show()

print('Silhouette Score:', silhouette_score(X, labels))
```

```python
inertias = []
k_values = range(1, 8)

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X)
    inertias.append(model.inertia_)

plt.figure(figsize=(6, 5))
plt.plot(k_values, inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
X_moons, _ = make_moons(n_samples=300, noise=0.08, random_state=42)

dbscan = DBSCAN(eps=0.25, min_samples=5)
db_labels = dbscan.fit_predict(X_moons)

plt.figure(figsize=(6, 5))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=db_labels, cmap='plasma')
plt.title('DBSCAN on Non-Linear Clusters')
plt.show()
```
