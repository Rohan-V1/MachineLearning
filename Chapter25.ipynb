# Self-Attention

Self-attention lets each token look at other tokens in the sequence and decide which tokens are important for understanding context.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import numpy as np
import pandas as pd
```

```python
sentence = 'machine learning is powerful'
tokens = sentence.split()

vocabulary = {word: i for i, word in enumerate(tokens)}

print('Tokens:', tokens)
print('Vocabulary:', vocabulary)
```

```python
rng = np.random.default_rng(42)
embedding_dim = 4
embeddings = rng.normal(size=(len(tokens), embedding_dim))

pd.DataFrame(embeddings, index=tokens)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()

query = embeddings[0]
scores = embeddings @ query
attention_weights = softmax(scores)

pd.DataFrame({
    'token': tokens,
    'attention_score': scores,
    'attention_weight': attention_weights
})
```

```python
context_vector = attention_weights @ embeddings
print('Context vector for token:', tokens[0])
print(context_vector)
```
