# Types of Machine Learning

Machine Learning algorithms are generally categorized based on how they learn from data.

## Main Types

```text
Machine Learning
│
├── Supervised Learning
│
├── Unsupervised Learning
│
├── Semi-Supervised Learning
│
├── Self-Supervised Learning
│
└── Reinforcement Learning
```

Each type has different applications, strengths, and limitations.

# 1. Supervised Learning

Supervised Learning is a type of Machine Learning where the model learns from **labeled data**.

### Definition

The dataset contains:

- Input Features (X)
- Correct Output Labels (Y)

The algorithm learns the relationship between X and Y.

### Example

| Hours Studied | Marks |
|--------------|--------|
| 2 | 30 |
| 4 | 50 |
| 6 | 70 |
| 8 | 90 |

The model learns:

```text
Hours Studied → Marks
```

and predicts marks for new students.

### Applications

- House Price Prediction
- Spam Detection
- Disease Diagnosis
- Credit Risk Analysis
- Weather Forecasting

# Types of Supervised Learning

## 1. Regression

Used when output is continuous.

Examples:

- House Price Prediction
- Temperature Prediction
- Sales Forecasting

Example:

```text
Input: House Features
Output: ₹45,00,000
```

### Popular Algorithms

- Linear Regression
- Polynomial Regression
- Decision Tree Regressor
- Random Forest Regressor

---

## 2. Classification

Used when output belongs to categories.

Examples:

- Spam / Not Spam
- Disease / No Disease
- Pass / Fail

Example:

```text
Input: Email
Output: Spam
```

### Popular Algorithms

- Logistic Regression
- Naive Bayes
- SVM
- Decision Tree
- Random Forest

````python
# Supervised Learning Example - Linear Regression

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[2],[4],[6],[8]])
y = np.array([30,50,70,90])

model = LinearRegression()

model.fit(X,y)

prediction = model.predict([[5]])

print("Predicted Marks:", prediction[0])
````

# Advantages of Supervised Learning

✅ High accuracy when sufficient labeled data exists

✅ Easy to evaluate performance

✅ Works well for prediction problems

✅ Widely used in industry

# Limitations

❌ Requires labeled data

❌ Labeling can be expensive

❌ May overfit training data

❌ Performance depends on data quality

# 2. Unsupervised Learning

Unsupervised Learning works with **unlabeled data**.

The algorithm discovers hidden patterns without knowing the correct outputs.

### Example

Customer purchase data:

| Customer | Age | Income |
|----------|------|---------|
| A | 25 | 30k |
| B | 26 | 32k |
| C | 50 | 90k |

The algorithm automatically groups similar customers.

### Applications

- Customer Segmentation
- Recommendation Systems
- Market Basket Analysis
- Anomaly Detection
- Pattern Recognition

# Types of Unsupervised Learning

## 1. Clustering

Groups similar data points together.

Example:

```text
Students grouped based on marks
```

Popular Algorithms:

- K-Means
- DBSCAN
- Hierarchical Clustering

---

## 2. Association

Discovers relationships between variables.

Example:

```text
People buying Bread often buy Butter
```

Algorithms:

- Apriori
- Eclat
- FP-Growth

---

## 3. Dimensionality Reduction

Reduces number of features.

Example:

```text
100 Features → 10 Features
```

Algorithms:

- PCA
- t-SNE
- LDA

````python
import warnings
warnings.filterwarnings("ignore")
# K-Means Clustering Example

from sklearn.cluster import KMeans
import numpy as np

X = np.array([
    [1,2],
    [1,4],
    [1,0],
    [10,2],
    [10,4],
    [10,0]
])

kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(X)

print("Cluster Labels:")
print(kmeans.labels_)
````

# Advantages of Unsupervised Learning

✅ No labeled data required

✅ Finds hidden patterns

✅ Useful for exploratory analysis

✅ Discovers unknown relationships

# Limitations

❌ Difficult to evaluate

❌ Results may be hard to interpret

❌ Sensitive to parameter selection

# 3. Semi-Supervised Learning

Semi-Supervised Learning combines:

- Small amount of labeled data
- Large amount of unlabeled data

### Why?

Labeling data is expensive.

Example:

```text
1000 Images

50 Labeled
950 Unlabeled
```

The model learns using both datasets.

### Applications

- Medical Imaging
- Speech Recognition
- Image Classification
- Web Content Classification

### Advantages

✅ Reduces labeling cost

✅ Better accuracy than unsupervised learning

✅ Efficient use of available data

# 4. Self-Supervised Learning

A modern learning technique where labels are generated automatically from data itself.

### Example

Given:

```text
"The cat sat on the ____"
```

Model learns to predict:

```text
mat
```

No human label is required.

### Applications

- ChatGPT
- BERT
- GPT Models
- Image Representation Learning

### Benefits

✅ Learns from massive datasets

✅ Reduces manual labeling

✅ Foundation for modern AI systems

# 5. Reinforcement Learning

Reinforcement Learning (RL) learns through interaction with an environment.

The model receives:

- Rewards
- Penalties

and learns the best actions.

### Components

Agent
Environment
State
Action
Reward

### Workflow

```text
Agent
  ↓ Action

Environment
  ↓ Reward

Agent learns
```

### Examples

- Chess AI
- Robotics
- Self-Driving Cars
- Game Playing

# Reinforcement Learning Example

Imagine a robot learning to walk.

### Correct Step

Reward = +10

### Wrong Step

Reward = -5

Over time:

```text
Good Actions → More Rewards

Bad Actions → More Penalties
```

The robot learns the best walking strategy.
