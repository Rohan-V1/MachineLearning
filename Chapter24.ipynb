# Python Implementation

The following cells show a small practical example for this chapter.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.array([
    [0, 0, 0, 0, 0, 0],
    [0, 255, 255, 255, 255, 0],
    [0, 255, 0, 0, 255, 0],
    [0, 255, 0, 0, 255, 0],
    [0, 255, 255, 255, 255, 0],
    [0, 0, 0, 0, 0, 0]
], dtype=np.uint8)

plt.imshow(image, cmap='gray')
plt.title('Simple Grayscale Image')
plt.axis('off')
plt.show()
```

```python
normalized_image = image / 255.0

print('Original min/max:', image.min(), image.max())
print('Normalized min/max:', normalized_image.min(), normalized_image.max())
```

```python
horizontal_edge_filter = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])

def convolve2d(matrix, kernel):
    h, w = matrix.shape
    kh, kw = kernel.shape
    output = np.zeros((h - kh + 1, w - kw + 1))

    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            output[i, j] = np.sum(matrix[i:i+kh, j:j+kw] * kernel)

    return output

edges = convolve2d(normalized_image, horizontal_edge_filter)

plt.imshow(edges, cmap='gray')
plt.title('Detected Horizontal Edges')
plt.axis('off')
plt.show()
```
