# MLOps

MLOps combines machine learning, software engineering, and operations.

It helps teams build, deploy, monitor, and maintain ML systems reliably.

# Why MLOps Matters

ML projects can fail after deployment because of:

- Data drift
- Model drift
- Poor reproducibility
- Missing monitoring
- Manual deployment steps
- Training-serving skew

# MLOps Lifecycle

1. Data versioning
2. Experiment tracking
3. Model training
4. Model registry
5. CI/CD
6. Deployment
7. Monitoring
8. Retraining

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import json
import time
from pathlib import Path
```

```python
experiment = {
    'model': 'LogisticRegression',
    'dataset': 'breast_cancer',
    'accuracy': 0.96,
    'created_at': time.strftime('%Y-%m-%d %H:%M:%S')
}

Path('chapter26_experiment_log.json').write_text(json.dumps(experiment, indent=2))
print('Experiment logged')
```

```python
loaded = json.loads(Path('chapter26_experiment_log.json').read_text())
loaded
```

```python
def check_data_drift(training_mean, production_mean, threshold=0.2):
    drift = abs(training_mean - production_mean)
    return drift > threshold, drift

has_drift, drift_value = check_data_drift(10.0, 10.35)

print('Drift detected:', has_drift)
print('Drift value:', drift_value)
```
