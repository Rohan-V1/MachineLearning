# Reinforcement Learning

Reinforcement Learning trains an agent to make decisions by interacting with an environment.

The agent learns through rewards and penalties.

# Key Terms

- **Agent:** learner or decision maker.
- **Environment:** world the agent interacts with.
- **State:** current situation.
- **Action:** choice made by the agent.
- **Reward:** feedback signal.
- **Policy:** strategy for choosing actions.

# Q-Learning

Q-Learning learns the value of taking an action in a state.

Update rule:

$$Q(s,a) = Q(s,a) + \alpha [r + \gamma \max Q(s',a') - Q(s,a)]$$

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import numpy as np
```

```python
# Simple 1D world: positions 0 to 4. Goal is position 4.
states = 5
actions = 2  # 0 = left, 1 = right
q_table = np.zeros((states, actions))

alpha = 0.1
gamma = 0.9
epsilon = 0.2
episodes = 200

rng = np.random.default_rng(42)
```

```python
for episode in range(episodes):
    state = 0

    while state != 4:
        if rng.random() < epsilon:
            action = rng.integers(actions)
        else:
            action = np.argmax(q_table[state])

        if action == 1:
            next_state = min(state + 1, 4)
        else:
            next_state = max(state - 1, 0)

        reward = 1 if next_state == 4 else 0

        q_table[state, action] = q_table[state, action] + alpha * (
            reward + gamma * np.max(q_table[next_state]) - q_table[state, action]
        )

        state = next_state

q_table
```

```python
state = 0
path = [state]

while state != 4:
    action = np.argmax(q_table[state])
    state = min(state + 1, 4) if action == 1 else max(state - 1, 0)
    path.append(state)

print('Learned path:', path)
```
