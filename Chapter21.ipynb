# Recommendation Systems

Recommendation Systems suggest items to users based on preferences, behavior, or similarity.

They are used by platforms such as e-commerce sites, streaming services, social media, and learning platforms.

# Types of Recommendation Systems

- **Popularity-based:** recommends popular items.
- **Content-based filtering:** recommends items similar to what the user liked.
- **Collaborative filtering:** recommends based on similar users or similar items.
- **Hybrid systems:** combine multiple approaches.

# Similarity

Recommendation systems often use similarity measures such as cosine similarity to compare users or items.

# Python Implementation

The following cells show a small practical example for this chapter.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.DataFrame({
    'User': ['A', 'A', 'A', 'B', 'B', 'C', 'C', 'D', 'D'],
    'Movie': ['M1', 'M2', 'M3', 'M1', 'M3', 'M2', 'M4', 'M1', 'M4'],
    'Rating': [5, 4, 1, 4, 2, 5, 4, 2, 5]
})

ratings
```

```python
user_item = ratings.pivot_table(index='User', columns='Movie', values='Rating').fillna(0)
user_item
```

```python
item_similarity = cosine_similarity(user_item.T)
item_similarity_df = pd.DataFrame(
    item_similarity,
    index=user_item.columns,
    columns=user_item.columns
)

item_similarity_df
```

```python
def recommend_similar_movies(movie, top_n=2):
    scores = item_similarity_df[movie].sort_values(ascending=False)
    return scores.drop(movie).head(top_n)

recommend_similar_movies('M1')
```
